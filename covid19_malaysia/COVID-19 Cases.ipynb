# COVID-19 Cases in Malaysia

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Daily recorded COVID-19 cases at country level

```python
url_cases1 = "https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/epidemic/cases_malaysia.csv"
```

```python
df1 = pd.read_csv(url_cases1,parse_dates=["date"],usecols=['date','cases_new','cases_recovered','cases_active'])
df1.set_index('date',inplace=True)
df1
```

```python
plt.figure(figsize=(15, 8))
plt.grid()
plt.title("Daily Recorded COVID-19 Cases")
plt.xlabel("Time")
plt.ylabel("Number of cases")
plt.plot(df1)
plt.legend(['cases_new','cases_import','cases_recovered','cases_active'])
```
